## Introdução e objetivos

### Introdução

O coração humano atua de maneira involuntária e autosustentável para manter o sangue circulando por todo o corpo. Seus batimentos são dados por uma sequência de oscilações e relaxamentos gerados pelas células marcapasso do próprio órgão. A taxa na qual o coração responde e reproduz esses impulsos não é constante, variando conforme a necessidade do corpo, seja em ambientes relaxados (como durante o sono) ou em resposta a estímulos externos e atividades físicas. O comportamento dos batimentos cardíacos pode ser comparado a um oscilador de relaxamento, permitindo modelar a atividade elétrica do coração como tal.

O modelo de FitzHugh-Nagumo, uma simplificação do modelo de Hodgkin-Huxley, oferece uma representação matemática dessa dinâmica elétrica. Esse modelo é especialmente útil para simular diferentes condições cardíacas ao ajustar seus parâmetros. Ao aplicar diferentes configurações de parâmetros, podemos reproduzir uma variedade de perfis clínicos, desde condições normais até diversas patologias cardíacas. Realizar testes com esses diferentes parâmetros possibilita a análise detalhada de como mudanças na dinâmica elétrica podem impactar o funcionamento cardíaco, fornecendo insights valiosos para diagnósticos e tratamentos.

# Modelo
O modelo utilizado é o sistema de Fitzhugh-Nagumo é descrito pelas seguintes equações diferenciais:

$\Large\frac{dx}{dt}$ = $ y + x - \frac{x^3}{3} + I(t)$

$\Large\frac{dy}{dt}$ = $ -x + a - by $

$ I(t) $ = $ Acos(\frac{1}{2}*t) $


Onde:
- $ x(t) $ representa o potencial de ação de um único miocárdio.
- $ y(t) $ representa o potencial de ação de um único marcapasso.
- $ I(t) $ é uma função periódica de forçamento.

- $ a $: Modula a não-linearidade da equação que descreve a variável de recuperação, afetando a forma e a propagação do potencial de ação.
Unidade de Medida: Adimensional, relacionado à dinâmica temporal do potencial de ação.
- $ b $: Ajusta a velocidade de recuperação do sistema após a ativação, afetando a taxa de repolarização das células cardíacas.
Unidade de Medida: Adimensional, envolvido na escala temporal relativa entre ativação e recuperação.
- $ A $: Representa a amplitude do estímulo externo aplicado, crucial para a resposta do potencial de ação.
Unidade de Medida:
tensão (mV).

- obs: "a" e "b": São constantes do sistema de equações diferenciais que modelam o comportamento do potencial de membrana e a ativação/inativação de canais iônicos no modelo Fitzhugh-Nagumo original.

Por se tratar de uma simplificação do modelo de Hodgkin-Huxley, muito fatores são simplificados nesse modelo devido à redução do sistema de quatro variáveis iniciais, sendo elas:

- Potencial de Membrana

- Ativação de Sódio

- Inativação do Sódio

- Ativação do Potássio

Mesmo apresentando um agrupamento mais aprofundado de dados a serem estudados, devido à dificuldade para visualizar e analisar uma amostra de quatro dimensões, o material de estudo do modelo foi reduzido para:

- Potencial de Membrana com Ativação de Sódio

- Inativação do Sódio com Ativação de Potássio

Outro fator que vale ser destacado é a adaptação de contexto do modelo, visto que foi inicialmente proposto para análise de neurônios ao invés de células cardíacas, evidenciando outro fator simplificador para o modelo.

Além disso, diversas questões do comportamento de células cardíacas não podem ser expostas por meio dele devido a essa simplificação, como cada complexidade quantitativa ou qualitativa da atividade elétrica celular, devido aos seguintes comportamentos inadequados:

- Corrente iônica (I(t)):

Durante o platô positivo do potencial de ação, os cátions fluem tanto para dentro (corrente positiva) quanto para fora (corrente negativa) das células, fazendo com que I(t) tivesse que ter um valor zero, entretanto não é o que acontece no modelo.

- Células Marca-passo:

Modelo sugere que x é síncrono com I (sistema sempre sob uma força externa, exceto nos pontos médios onde I = 0), contradizendo com o comportamento das células marca-passo (despolarizam sem impulsos externos), não explicando que pode ser regulado com relógios biológicos que controlam a atividade dos canais iônicos dessas células.

- Representação Quantitativa das variáveis x e y:

As variáveis x e y não representam exatamente o potencial de membrana e o fluxo de cátions nas células. Dessa forma, o modelo foi proetado para mostrar oscilações de relaxamento e um ciclo semelhante ao de Hodgkin-Huxley, limitando a capacidade de explicar a complexidade da atividade elétrica celular.

Obs. o modelo e equações utilizadas são limitados a apenas uma célula (de cada tipo) o que gera importantes insights da eletrofisiologia cardíaca, entretanto, não se pode afirmar com certeza se anomalias no comportamento de uma célula afetam todo o coração. Por exemplo, células de apenas uma região do tecido cardíaco podem ser afetadas, ainda assim, essas anomalias podem indicar sinais de "problemas".   

# Resultados e Discussão

Nesta seção, serão apresentados os resultados preliminares obtidos com o modelo simplificado. As simulações serão realizadas utilizando Python.

O primeiro passo é replicar os resultados obtidos no artigo. Foram utilizados como parâmetros a = 0.6, b = 0.75, A = -0.9. Após serão demonstrados os 3 perfis clínicos com seus respectivos parâmetros.

### Visualização

```python
#bibliotecas utilizadas
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Definindo o sistema de equações diferenciais
def fitz_nagumo(y0, t, a, b, A):
    x, y = y0
    dxdt = y + x - (x**3) / 3 + A * np.cos(0.5 * t)
    dydt = -x + a - b * y
    return [dxdt, dydt]

# Parâmetros a, b e A
a = 0.6
b = 0.75
A = -0.9

# Diferentes condições iniciais
initial_conditions = [
    [0, 0],
    [1, 0],
    [2, 0],
    [-1, 0]
]

# Vetor de tempo: de 0 a 100 segundos
t = np.linspace(0, 100, 500)

# Cores para os gráficos
colors = ['red', 'blue', 'green', 'purple']

# Plot dos resultados
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# Gráfico para x(t) e y(t) com diferentes condições iniciais
for i, y0 in enumerate(initial_conditions):
    sol = odeint(fitz_nagumo, y0, t, args=(a, b, A))
    axs[0].plot(t, sol[:, 0], label=f'x(t) - Condição {i+1}', color=colors[i])
    axs[1].plot(t, sol[:, 1], label=f'y(t) - Condição {i+1}', color=colors[i])
    axs[2].plot(sol[:, 0], sol[:, 1], label=f'Trajetória {i+1}', color=colors[i])

# Configurações do gráfico x(t)
axs[0].set_xlabel('Tempo')
axs[0].set_ylabel('x(t)')
axs[0].set_title('x(t) ao longo do tempo')
axs[0].set_xlim(0, 100)

# Configurações do gráfico y(t)
axs[1].set_xlabel('Tempo')
axs[1].set_ylabel('y(t)')
axs[1].set_title('y(t) ao longo do tempo')
axs[1].set_xlim(0, 100)

# Configurações do gráfico y por x
axs[2].set_xlabel('x(t)')
axs[2].set_ylabel('y(t)')
axs[2].legend()
axs[2].set_title('Plano de fase: y(t) vs x(t)')

plt.tight_layout()
plt.show()
```

```python
# Parâmetros para diferentes condições clínicas
conditions = {
    "Saudável": {'a': 0.7, 'b': 0.8, 'A': 1.0},
    "Fibrilação Atrial (FA)": {'a': 1, 'b': 0.2, 'A': 1},
    "Hipercalemia": {'a': 1.2, 'b': 0.1, 'A': 1.15}
}

# Condição inicial
initial_conditions = [0, 0]

# Vetor de tempo: de 0 a 100 segundos
t = np.linspace(0, 100, 500)  # 500 pontos

# Configurar o plot
fig, axs = plt.subplots(3, 2, figsize=(12, 10))

# Resolver e plotar para cada condição clínica
for i, (condition, params) in enumerate(conditions.items()):
    a, b, A = params['a'], params['b'], params['A']
    sol = odeint(fitz_nagumo, initial_conditions, t, args=(a, b, A))

    # Gráfico para x(t)
    axs[i, 0].plot(t, sol[:, 0], label=f'x(t) - {condition}')
    axs[i, 0].set_xlabel('Tempo')
    axs[i, 0].set_ylabel('x(t)')
    axs[i, 0].set_title(f'x(t) ao longo do tempo - {condition}')
    axs[i, 0].set_xlim(0, 100)

    # Gráfico para y(t)
    axs[i, 1].plot(t, sol[:, 1], label=f'y(t) - {condition}')
    axs[i, 1].set_xlabel('Tempo')
    axs[i, 1].set_ylabel('y(t)')
    axs[i, 1].set_title(f'y(t) ao longo do tempo - {condition}')
    axs[i, 1].set_xlim(0, 100)

plt.tight_layout()
plt.show()
```

O gráfico representando o coração de um usuário saudável apresenta um padrão regular. Após a despolarização de uma célula de marcapasso (entrada de Na+), ocorre a polarização de uma célula miocárdica. O potencial de ação tem uma grande subida, seguido por uma leve queda e um platô (entrada de íons de cálcio), e então uma queda vertiginosa na sua repolarização. Isso resulta em uma onda de potencial de ação consistente e regular ao longo do tempo, como mostrado nos gráficos de x(t) e y(t) ao longo do tempo para o estado saudável.

No caso de um paciente com fibrilação atrial (FA), o gráfico mostra uma atividade elétrica irregular. Embora a frequência dos ciclos de despolarização e repolarização possa parecer elevada, há divergências significativas na amplitude e na forma das ondas de potencial de ação. Essas alterações podem estar associadas a variações no potencial de membrana, que geram novos potenciais de ação e, consequentemente, eventos arrítmicos. Essas variações podem resultar em pós-potenciais precoces, ou mesmo após a repolarização, gerando pós-potenciais tardios. Esse comportamento é indicativo de uma dinâmica complexa nos processos de efluxo e influxo de íons, que influencia diretamente a regularidade e a estabilidade do ritmo cardíaco observado no gráfico.

Em um paciente com hipercalemia, os gráficos, utilizando como referência o apresentado acima, mostram um comportamento alterado devido ao excesso de potássio no sangue, que afeta a excitação e a recuperação das células cardíacas. Na hipercalemia, o potencial de ação das células marcapasso pode apresentar uma subida mais lenta e um platô menos pronunciado, refletindo a diminuição na diferença de potencial elétrico durante a despolarização. Isso leva a uma diminuição da amplitude das ondas de potencial de ação e a um padrão mais irregular de despolarização e repolarização. A repolarização pode ser prolongada, resultando em um potencial de ação com picos mais baixos e uma atividade elétrica mais desorganizada.

Diferentemente da fibrilação atrial, que se caracteriza por uma atividade elétrica caótica e irregular, a hipercalemia provoca uma diminuição na amplitude e uma irregularidade no padrão de despolarização e repolarização, mas sem necessariamente aumentar a frequência dos ciclos cardíacos. Os gráficos de x(t) e y(t) ao longo do tempo para a hipercalemia mostram essa tendência de uma atividade elétrica menos intensa e mais errática, refletindo as alterações causadas pelo excesso de potássio no sistema cardíaco.
